```python
import numpy as np
import pandas as pd 
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
train1 = pd.read_table('data/Train1.txt', delim_whitespace=True, names=('dimension_1', 'dimension_2'))
train2 = pd.read_table('data/Train2.txt', delim_whitespace=True, names=('dimension_1', 'dimension_2'))
original_data = pd.concat([train1, train2])
# read the Testing data files
test1 = pd.read_table('data/Test1.txt', delim_whitespace=True, names=('dimension_1', 'dimension_2'))
test2 = pd.read_table('data/Test2.txt', delim_whitespace=True, names=('dimension_1', 'dimension_2'))
# Training and Validation Data
train_data = pd.concat([train1[: 1500], train2[: 1500]])
#Mean and Standard Deviation of the Training data
train_data_mean = train_data.mean()
train_data_std = train_data.std()
#Normalized Train data
train_data_normalized = (train_data - train_data.mean())/train_data.std()
# Normalized Test Data
test_data = (pd.concat([test1, test2]) - train_data_mean)/train_data_std
```

```python
train_data_target = int(len(train_data)/2) * [0] + int(len(train_data)/2)* [1]
validation_data = (pd.concat([train1[1500: ], train2[1500:]]) - train_data_mean)/train_data_std
validation_data_target = int(len(validation_data)/2)*[0] + int(len(validation_data)/2)*[1]
test_data_target = int(len(test1)) * [0] + int(len(test2)) * [1]
```

```python
X = train_data_normalized.to_numpy()
X = X.T
X_validation = validation_data.to_numpy()
X_validation = X_validation.T
hidden_layer_neurons = [2, 4, 6, 8, 10]
input_layer_neurons = X.shape[0]
output_layer_neuron = 1
y_class = []
```

```python
def non_linear_activation(x):
    return 1/(1+ np.exp(-x))
```

```python
alpha = 0.0001
```

```python
def multilayer_perceptron(nh):
    W = np.random.uniform(size=(input_layer_neurons, nh))
    W_hat = np.random.uniform(size=(nh, output_layer_neuron))
    validation_loss_prev_value = 50
    losses = []
    validation_losses = []
    test_losses = []
    validation_loss_value = 50
    for epoch in range(1000000):
        validation_loss_prev_value = validation_loss_value

        #Forward Propagation
        h = np.dot(W.T, X)
        h_activation = non_linear_activation(h) 
        o = np.dot(W_hat.T, h_activation)
        y = non_linear_activation(o)

        #Backward Propagation
        #Training Error
        e = np.square(train_data_target - y)/2
        de_wrt_o = -(train_data_target-y)
        dy_wrt_o = np.multiply(y, (1-y))
        #Rate of change of error wrt W-> hidden and output layer
        e_hat = np.dot(h_activation, (de_wrt_o*dy_wrt_o).T)

        #Rate of change of error W_hat -> inupt and hidden layer
        dh_activation_wrt_h = np.multiply(h_activation, (1 - h_activation))
        de_wrt_W = np.dot(X, (dh_activation_wrt_h * (np.dot(W_hat, (dy_wrt_o*de_wrt_o)))).T)

        # Weight update based on Batch Gradientt descent
        W_hat = W_hat - alpha* e_hat
        W = W - alpha* de_wrt_W

        #Loss on validation set
        validation_h = non_linear_activation(np.dot(W.T , X_validation))
        validation_y =  non_linear_activation(np.dot(W_hat.T, validation_h))
        validation_loss = np.square(validation_data_target - validation_y)/2
        validation_loss_value = np.average(validation_loss)
        validation_losses.append(validation_loss_value)
        
        test_h = non_linear_activation(np.dot(W.T, X_test))
        test_y = non_linear_activation(np.dot(W_hat.T, test_h))
        test_loss = np.square(test_data_target - test_y)/2
        test_loss_value = np.average(test_loss)
        test_losses.append(test_loss_value)
        
        if validation_loss_prev_value < validation_loss_value:
            break

        epoch_loss = np.average(e)
        if epoch % 100 == 0:
            print(f"Error at Epoch {epoch} is {epoch_loss:.5f}")
    
        losses.append(epoch_loss)
    return W, W_hat, losses, validation_losses,test_losses, epoch
    
```

```python
W, W_hat, loss, validation_loss, test_loss, epoch = multilayer_perceptron(8)
```

```python
X_test = test_data.to_numpy()
X_test = X_test.T
# prediction
h_predicted = non_linear_activation(np.dot(W.T , X_test))
y_predicted = non_linear_activation(np.dot(W_hat.T, h_predicted))
y_predicted = (y_predicted.T).tolist()
print(len(y_predicted))
print(type(y_predicted))
y_class = []
#Class prediction for sigmoid function
for x in y_predicted:
    #print(x)
    if x[0] > 0.5:
        y_class.append(1)
    elif x[0] < 0.5:
        y_class.append(0)
    else:
        y_class.append(np.random([0,1]))
#print(y_class)
print(accuracy_score(y_class, test_data_target))
```

```python
e = [i for i in range(epoch)]
e1 = [i for i in range(epoch +1)]
```
